```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('book.csv',encoding= 'Latin1')
data
```

# Data Preprocessing

```python
data_new = data.drop(['Unnamed: 0'],axis=1)
data_new
```

```python
# Sort by User IDs
data_new.sort_values(['User.ID'])
```

```python
# Number of unique users in the dataset
len(data_new['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(data_new['Book.Title'].unique())
```

```python
data1 = data_new.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index(drop=True)
data1
```

```python
data1.index = data_new['User.ID'].unique()
data1.index
```

```python
data1.fillna(0, inplace=True)
```

```python
data1
```

# Visualization

```python
import seaborn as sns
```

```python
sns.countplot("Book.Rating",data=data)
```

# Scatterplot between Book.Rating and User.ID

```python
sns.scatterplot("Book.Rating","User.ID",data=data)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1 - pairwise_distances(data1.values,metric='cosine')
user_sim
```

```python
sns.heatmap(user_sim,cmap="viridis")
```

```python
# Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
# Set the index and column names to user ids
user_sim_df.index = data_new['User.ID'].unique()
user_sim_df.columns=data_new['User.ID'].unique()
user_sim_df
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim_df
```

```python
# Most similar users
user_sim_df.idxmax(axis=1)
```
